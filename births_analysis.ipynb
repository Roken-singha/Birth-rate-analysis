```python
import pandas as pd
import matplotlib.pyplot as plt

# Load and clean data
df = pd.read_csv("births.csv")
df_clean = df.dropna(subset=['day'])
df_clean['day'] = df_clean['day'].astype(int)

# Group data by year and sum births
yearly_births = df_clean.groupby('year')['births'].sum()

# Plot total births per year
plt.figure(figsize=(8, 4))
yearly_births.plot()
plt.title("Total Births per Year")
plt.xlabel("Year")
plt.ylabel("Number of Births")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load and clean data
df = pd.read_csv("births.csv")
df_clean = df.dropna(subset=['day'])
df_clean['day'] = df_clean['day'].astype(int)

# Group data by gender and sum births
gender_births = df_clean.groupby('gender')['births'].sum()

# Plot total births by gender
plt.figure(figsize=(6, 4))
gender_births.plot(kind='bar', color=['pink', 'lightblue'])
plt.title("Total Births by Gender")
plt.xlabel("Gender")
plt.ylabel("Number of Births")
plt.grid(axis='y')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load and clean data
df = pd.read_csv("births.csv")
df_clean = df.dropna(subset=['day'])
df_clean['day'] = df_clean['day'].astype(int)

# Create pivot table of average births by day and month
pivot = df_clean.pivot_table(index='day', columns='month', values='births', aggfunc='mean')

# Plot heatmap using matplotlib
fig, ax = plt.subplots(figsize=(10, 6))
c = ax.imshow(pivot, cmap='coolwarm', aspect='auto')

fig.colorbar(c, ax=ax)
ax.set_title("Average Daily Births by Day and Month")
ax.set_xlabel("Month")
ax.set_ylabel("Day")
ax.set_xticks(range(len(pivot.columns)))
ax.set_xticklabels(pivot.columns)
ax.set_yticks(range(len(pivot.index)))
ax.set_yticklabels(pivot.index)
plt.tight_layout()
plt.show()
```
